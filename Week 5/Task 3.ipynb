```python
import os, base64, json
from mistralai import Mistral

client = Mistral(api_key=mistral_key)

def encode_image(path):
    return base64.b64encode(open(path, "rb").read()).decode()

response = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "image_url",
        "image_url": f"data:image/jpeg;base64,{encode_image('UAE_id.jpg')}"
    },
    include_image_base64=True
)

# print(json.dumps(response.model_dump(), indent=2))
```

```python
# Process Mistral OCR output to separate Arabic and English lines
import json

markdown_text = response.pages[0].markdown

arabic_lines = []
english_lines = []

for line in markdown_text.split("\n"):
    text = line.strip("**").strip()
    if not text:
        continue
    if any('\u0600' <= c <= '\u06FF' for c in text):
        arabic_lines.append(text)
    else:
        english_lines.append(text)

print(" Arabic lines:", arabic_lines)
print(" English lines:", english_lines)
```

```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display

# Load image
image = cv2.imread(IMAGE_PATH)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Create a white canvas with same size
white_canvas = np.ones_like(image_rgb) * 255

# Convert to PIL for Arabic text rendering
pil_canvas = Image.fromarray(white_canvas)
draw = ImageDraw.Draw(pil_canvas)

# Load Arabic-supporting font
font = ImageFont.truetype("arial.ttf", 16)  # Adjust size as needed

# Draw detections on original image
for line in predictions[0].text_lines:
    pts = np.array(line.polygon, np.int32).reshape((-1, 1, 2))
    cv2.polylines(image_rgb, [pts], isClosed=True, color=(255, 0, 0), thickness=2)

# Draw OCR text on white canvas
for line in predictions[0].text_lines:
    reshaped_text = arabic_reshaper.reshape(line.text)
    bidi_text = get_display(reshaped_text)

    x, y = int(line.polygon[0][0]), int(line.polygon[0][1])
    draw.text((x, y), bidi_text, font=font, fill=(0, 0, 0))

# Convert PIL canvas back to numpy BGR for OpenCV display
white_canvas_bgr = cv2.cvtColor(np.array(pil_canvas), cv2.COLOR_RGB2BGR)

# Display both images in separate cv2 windows
cv2.imshow("Detections", cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))
cv2.imshow("OCR Text Overlay", white_canvas_bgr)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Assuming layout_predictions is your provided list
layout_results = layout_predictions[0].bboxes

# Load image
image = cv2.imread(IMAGE_PATH)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Draw layout boxes
for box in layout_results:
    # Extract polygon points
    pts = np.array(box.polygon, np.int32).reshape((-1, 1, 2))
    
    # Draw polygon
    cv2.polylines(image_rgb, [pts], isClosed=True, color=(0, 255, 0), thickness=2)
    
    # Write label and confidence
    label_text = f"{box.label} ({box.confidence:.2f})"
    
    # Get top-left corner for placing label
    x, y = int(box.polygon[0][0]), int(box.polygon[0][1]) - 10
    
    # Draw background rectangle for better visibility
    (text_w, text_h), baseline = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
    cv2.rectangle(image_rgb, (x, y - text_h - baseline), (x + text_w, y + baseline), (0, 255, 0), -1)
    
    # Put label text
    cv2.putText(image_rgb, label_text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)

# Convert to BGR for cv2 display or save
image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)

# Display using cv2 (for scripts)
cv2.imshow("Layout Predictions", image_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Or for Gradio, return image_rgb directly
```

### Gradio

```python
import gradio as gr
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
from surya.recognition import RecognitionPredictor
from surya.detection import DetectionPredictor
from surya.layout import LayoutPredictor 

# Initialize predictors once
det_predictor = DetectionPredictor()
rec_predictor = RecognitionPredictor()
layout_predictor = LayoutPredictor()

# Main processing function
def process_image(image_pil):
    # Convert to OpenCV format
    image_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)
    original_image = image_cv.copy()

    ### --- Layout Predictions ---
    layout_predictions = layout_predictor([image_pil])  # Adjust based on your API

    layout_image = original_image.copy()
    for box in layout_predictions[0].bboxes:
        pts = np.array(box.polygon, np.int32).reshape((-1, 1, 2))
        cv2.polylines(layout_image, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

        label_text = f"{box.label} ({box.confidence:.2f})"
        x, y = int(box.polygon[0][0]), int(box.polygon[0][1]) - 10
        (text_w, text_h), baseline = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
        cv2.rectangle(layout_image, (x, y - text_h - baseline), (x + text_w, y + baseline), (0, 255, 0), -1)
        cv2.putText(layout_image, label_text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)

    ### --- Detection Predictions (OCR boxes + text) ---
    detection_image = original_image.copy()
    predictions = rec_predictor([image_pil], det_predictor=det_predictor)

    for line in predictions[0].text_lines:
        pts = np.array(line.polygon, np.int32).reshape((-1, 1, 2))
        cv2.polylines(detection_image, [pts], isClosed=True, color=(0, 0, 255), thickness=2)

        reshaped_text = arabic_reshaper.reshape(line.text)
        bidi_text = get_display(reshaped_text)

        image_for_pil = Image.fromarray(cv2.cvtColor(detection_image, cv2.COLOR_BGR2RGB))
        draw = ImageDraw.Draw(image_for_pil)
        font = ImageFont.truetype("arial.ttf", 16)
        x, y = int(line.polygon[0][0]), int(line.polygon[0][1])
        draw.text((x, y - 25), bidi_text, font=font, fill=(255, 0, 0))
        detection_image = cv2.cvtColor(np.array(image_for_pil), cv2.COLOR_RGB2BGR)

    ### --- OCR Text Output ---
    text_output = ""
    for line in predictions[0].text_lines:
        text_output += f"Text Line: {line.text}\n"
        text_output += f"Confidence: {line.confidence:.2f}\n"
        text_output += "-" * 50 + "\n"

    ### --- Convert images to RGB for Gradio display ---
    original_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    layout_rgb = cv2.cvtColor(layout_image, cv2.COLOR_BGR2RGB)
    detection_rgb = cv2.cvtColor(detection_image, cv2.COLOR_BGR2RGB)

    return layout_rgb, detection_rgb, text_output

# Gradio Interface
iface = gr.Interface(
    fn=process_image,
    inputs=gr.Image(type="pil"),
    outputs=[
        gr.Image(type="numpy", label="Layout Predictions"),
        gr.Image(type="numpy", label="Detection Predictions"),
        gr.Textbox(label="OCR Text Output")
    ],
    title="OCR & Layout Viewer with Surya",
    description="Upload an image to view original, layout predictions, detection predictions, and extracted OCR text."
)

iface.launch()
```
